## A trivial end-to-end check

Build a tiny Pandas DataFrame, then have DuckDB query it directly with SQL.
This is the core trick we'll lean on throughout the project: DuckDB can run
SQL directly against Pandas DataFrames (and CSV/Parquet files) with no server,
no import step, and no schema setup — it just works against a file on disk.

```python
df = pd.DataFrame({
    "council_district": [1, 2, 3, 4],
    "open_requests": [12, 47, 8, 23],
})

duckdb.sql("SELECT * FROM df WHERE open_requests > 10 ORDER BY open_requests DESC")
```
